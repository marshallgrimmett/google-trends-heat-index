### Imports and user-defined functions

```python
import pandas as pd
import numpy as np
from pytrends.request import TrendReq
from pytrends.exceptions import ResponseError
from requests.exceptions import HTTPError, ConnectionError
from datetime import datetime, timedelta
import time
from IPython.display import clear_output
import requests

def printProgress(progress, i, nonEmptycount, avgTime, timeRemain, totalTime, parts, errCount, err):
    clear_output(wait=True)
    print('Progress: ' + str(progress) + '% || ' +
          'Number of batches processed: ' + str(i) + ' || ' +
          'Number of batches with data: ' + str(nonEmptycount) + ' || ' +
          'Estimated time: ' + str(avgTime) + ' seconds || ' +
          'Time elapsed: ' + str(totalTime) + ' seconds || ' +
          'Time remaining: ' + str(timeRemain) + ' seconds || ' +
          'Parts: ' + str(parts) + ' || ' +
          'Error count: ' + str(errCount) + ' || ' +
          'Error: ' + str(err))

def readCypress():
    raw = pd.read_excel('data\\Cypress Q2\'20 Heat Index.xlsx', sheet_name='Sheet1')

    rawFiltered = raw.loc[:, ['Ordering MPN', 'Heat']]
    rawFiltered.drop_duplicates('Ordering MPN', keep='first', inplace=True)

    X = rawFiltered.loc[:, 'Ordering MPN']     # mpn to search for
    Y = rawFiltered.loc[:, 'Heat']    # manually entered heat

    X = X.tolist()
    Y = Y.tolist()

    # 157 Hot
    # 3 Hot
    # 221 Warm
    # 1676 Cold
    
    return X, Y

def readSkyHigh():
    raw = pd.read_excel('data\\6_3 Sky High Q2 Analysis.xlsx', sheet_name='Bidding_summary')

    rawFiltered = raw.loc[:, ['MPN', 'Heat']]
    rawFiltered.drop_duplicates('MPN', keep='first', inplace=True)

    X = rawFiltered.loc[:, 'MPN']     # mpn to search for
    Y = rawFiltered.loc[:, 'Heat']    # manually entered heat

    X = X.tolist()
    Y = Y.tolist()
    
    return X, Y

def readParts():
    raw = pd.read_excel('data\\45292-parts.xlsx', sheet_name='Sheet1')
    X = raw['Prcpart'].tolist()
    return X
```

### Main Script

```python
# dataName = 'sky_high'
# dataName = 'cypress'
dataName = 'all'

# Read in data
if dataName == 'sky_high':
    X, Y = readSkyHigh()
elif dataName == 'cypress':
    X, Y = readCypress()
else:
    X = readParts()

# Optional: select unfinished records. In case we hit rate limit or other errors occur.
X = X[5791:]

# if the part starts with "CON" or "STK" remove it
# Also remove any characters that cause bad requests
for i, part in enumerate(X):
    if part and (len(part) > 2) and ((part[:3].lower() == 'con') or (part[:3].lower() == 'stk')):
        part = part[3:]
        X[i] = part
    if part and (part[0] == '-'):
        part = part[1:]
        X[i] = part
    if part and (part[len(part)-1] == '-'):
        part = part[:len(part)-1]
        X[i] = part
    if '-+' in part:
        part = part[:part.index('-+')] + part[part.index('-+')+2:]
        X[i] = part

# Remove empty strings that cause bad requests
X = list(filter(None, X))

# pytrends package uses google trends api
pytrends = TrendReq(hl='en-US', tz=360)

# If blocked by google, uses proxies to bypass
# pytrends = TrendReq(hl='en-US', tz=360, timeout=(10,25), proxies=['https://34.203.233.13:80',], retries=2, backoff_factor=0.1, requests_args={'verify':False})

# Time frame to search
days = 365 * 2
timeframe = datetime.today() - timedelta(days=days)
timeframe = timeframe.strftime('%Y-%m-%d') + ' ' + datetime.today().strftime('%Y-%m-%d')

# Final dataframe to store all the google trends data
finalDf = pd.DataFrame()

# Used for computing timers and progress trackers
startTime = time.time()
totalTime = 0
sleepTime = 12
estimatedTime = []
progress = 0
avgTime = 0
timeRemain = 0

log = []

# count of non empty returned requests.
nonEmptycount = 0

# Split X into batches of size n to be sent per request (max of 5 currently)
n = 5
Xchunks = [X[i * n:(i + 1) * n] for i in range((len(X) + n - 1) // n)]

# For each part, search for its popularity
# Note: Numbers represent search interest relative to the highest point on the chart for the given region and time.
#       A value of 100 is the peak popularity for the term.
#       A value of 50 means that the term is half as popular.
#       A score of 0 means there was not enough data for this term.
for i, parts in enumerate(Xchunks):

    # record the start of each request
    begin_time = time.time()

    while True:
        try:
            # build the search payload
            pytrends.build_payload(
                 kw_list=parts,
                 cat=0,
                 timeframe=timeframe,
                 geo='',
                 gprop='')
            
            # compute popularity
            data = pytrends.interest_over_time()
        except Exception as e:
            log.append(e)
            printProgress(progress, (i+1), nonEmptycount, avgTime, timeRemain, totalTime, parts, len(log), 'Yes')
            time.sleep(60)
            continue
        break

    # Drop isPartial, not sure what it is
    if 'isPartial' in data.columns:
        data = data.drop(labels=['isPartial'], axis='columns')

    # append columns to final DF
    if not data.empty:
        nonEmptycount += 1
        finalDf[parts] = data[parts]

    # Pause between requests so we don't exceed google's rate limit
    time.sleep(sleepTime)

    # Print out progress trackers
    progress = round((i+1)/len(Xchunks)*100, 2)
    estimatedTime.append(round((len(Xchunks) * (time.time() - begin_time + sleepTime)), 2))
    avgTime = round((sum(estimatedTime) / len(estimatedTime)), 2)
    totalTime = round((totalTime + time.time() - begin_time + sleepTime), 2)
    timeRemain = round((avgTime - totalTime), 2)
    printProgress(progress, (i+1), nonEmptycount, avgTime, timeRemain, totalTime, parts, len(log), 'No')

# Save
saveFile = dataName + '_google_trends_popularity2.csv'
finalDf.to_csv(saveFile, encoding='utf_8_sig')

# Progress: 52.19% || Number of batches processed: 1158 || Number of batches with data: 207 || Estimated time: 59182.01 seconds || Time elapsed: 30885.66 seconds || Time remaining: 28296.35 seconds || Parts: ['H8-1K15-0.5%-50ppm', 'H8-1K18-0.1%-50ppm', 'H8-1K25-0.1%-50ppm', 'H8-1K33-0.1%-15ppm', 'H8-1K4-0.1%-15ppm'] || Error count: 175 || Error: Yes

print("\n--- %s seconds ---" % round(time.time() - startTime, 2))
print('Done.')
```

### Compute Accuracy

```python
# Used for finding thresholds
sumPopularity = []
meanPopularity = []

# Thresholds for determining Heat
threshold1 = 10
threshold2 = 100

# Store the computed Heat values
Y_pred = []

# Compute the heat for each part
for i, part in enumerate(X):
    
    # Conditional determines Heat based on the total sum of interest over the time period.
    if (part not in finalDf) or (finalDf[part].sum(axis=0, skipna=True) < threshold1):
        Y_pred.append('Cold')
    elif finalDf[part].sum(axis=0, skipna=True) < threshold2:
        Y_pred.append('Warm')
    else:
        Y_pred.append('Hot')
    
    # Compute Sum and mean values for finding thresholds
    if part in finalDf:
        meanPopularity.append(finalDf[part].mean(axis=0))
        sumPopularity.append(finalDf[part].sum(axis=0, skipna=True))
    else:
        sumPopularity.append(0)
        meanPopularity.append(0)

print([i for i in sumPopularity if i > 0])
print([i for i in meanPopularity if i > 0])
```

```python
# Accuracy to manually entered data
count = 0
for i in range(len(Y)):
    if Y[i].lower() == Y_pred[i].lower():
        count += 1

accuracy = count / len(Y)
print(str(round(accuracy*100, 2)) + '%')
```
